```python
import requests
from bs4 import BeautifulSoup
from bs4.element import NavigableString
import csv
import re
import pandas as pd

url = 'https://dsc-capstone.org/enrollment/'

response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.text, 'html.parser')

domain_sections = soup.find_all("div", class_="staffer")

offerings = []

for class_ in domain_sections:
    offering = {}

    section = class_.find("h3", class_="staffer-name")

    # if section.get("class") != ["staffer-name"]:
    #     print(section.get("class"))
    #     continue

    offering["title"] = section.contents[2].get_text(strip=True)
    offering["people"] = section.contents[4].get_text(strip=True).split("‚Ä¢")
    offering["location"] = section.contents[7].get_text(strip=True).replace("TA:", "")
    offering["section"] = section.contents[10].get_text(strip=True)
    offering["seats"] = int(section.contents[12].get_text(strip=True).replace(" seats", ""))
    offering["time"] = section.contents[14].get_text(strip=True)

    for sibling in section.next_siblings:
        if isinstance(sibling, NavigableString) and sibling.strip("\n"):
            offering["description"] = sibling.strip("\n\t ")

    details = class_.find_all("li")

    offering["about"] = details[0].get_text(strip=True).strip("About:")
    offering["mentoring_style"] = details[1].get_text(strip=True)
    # offering["suggested_prerequisites"] = details[2].get_text(strip=True).strip("Suggested Prerequisites:")
    offering["summer_tasks"] = details[3].get_text(strip=True).replace("Summer Tasks:", "")
    # offering["previous_projects"] = details[4].get_text(strip=True).strip("Previous Projects:")

    offerings.append(offering)

df = pd.DataFrame(offerings)
display(df)

df.to_csv('dsc_capstone_offerings.csv', index=False)
print("Data has been successfully written to 'dsc_capstone_offerings.csv'.")
```
